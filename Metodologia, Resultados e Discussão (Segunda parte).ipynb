```python
# Importando biblioteca
import pandas as pd
import pandas_datareader as web
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import yfinance as yf
```

```python
# Conexão com os dados do Yahoo Finance
data = web.get_data_yahoo('BRL=X')
```

```python
# Exibir as 5 primeiras linhas
data.head(5)
```

```python
# Criar um novo dataframe apenas com as colunas necessárias para treinar o modelo
features = ['Open', 'High', 'Low']
treino = data[features]
treino.reset_index()

# Criar um novo dataframe com os valores de fechmaneto real
y = data['Close']
```

```python
# Divisão dos dados entre treino/teste
X_treino, X_teste, y_treino, y_teste = train_test_split(treino, y, random_state=42)
```

```python
# Criar o modelo
lr_model = LinearRegression()
lr_model.fit(X_treino, y_treino)

# Predizer valores
predicao = lr_model.predict(X_teste)[:10]
```

```python
# Criar duas listas para armazenar o valor Real x Predito
lista_real = []
lista_predicao = []
for i in data['Close'][:10]:
    i = round(i, 2)
    lista_real.append(i)
for i in predicao:
    i = round(i, 2)
    lista_predicao.append(i)
```

```python
# Criar um dicionário com os valores
dict_regressao = {
    
    'Real':lista_real,
    'Predito':lista_predicao
    
}
```

```python
# Criar um dataframe novo para comparar os preços
df = pd.DataFrame(dict_regressao, columns=['Real', 'Predito'])

# Criar uma coluna com a difereça do preço Real x Predito
df['Erro'] = df['Real'] - df['Predito']
```
