FOREX - Previsão de preço do Dólar

Forex é um mercado financeiro descentralizado destinado a transações de câmbio, sendo o maior mercado do mundo. Em termos de volume de dinheiro, movimenta o equivalente a mais de 5 trilhões de dólares americanos diariamente. Segundo dados de 2007, movimentava cerca de 3,43 vezes mais do que a soma de todos os mercados de títulos no mundo, e 9,63 vezes o volume negociado no mercado de ações mundial. Estando ciente do alto volume negociado e a possibilidade de operar 24hrs por dia, este projeto se propõe a realizar a previsão do preço do dólar através de métodos de aprendizado de máquina.

Objetivo

Esse projeto tem como objetivo aplicar algoritmos de machine learning para previsão de preço no mercado financeiro.

Especificação Técnica

Dataset: Será utilizado os dados históricos do dólar x reais, disponível na biblioteca do Yahoo Finance no Python!

Formato: A base de dados está em um dataframe, contendo sete colunas (date, high, low, open, close, volume, adj close).

Métodos de Pré-processamento: Limpeza da base, onde serão removidos as colunas que não serão utilizadas no modelo.

Tarefa de Aprendizado: Será aplicado a tarefa de Regressão Linear

Algoritmos Avaliados: Árvore de Decisão
